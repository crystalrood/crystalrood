```python
df = mat_leave [[0,1]]
df.columns = ['Company', 'Paid_Weeks']
#sort companies in asc order
df.head()
```

```python
#group companies and sort in descending order
ranked_companies = df.groupby('Company').sum().sort_values(by='Paid_Weeks',ascending=False).reset_index()
ranked_companies.head()
#complete table list - https://plot.ly/python/table/
```

```python
#[[0, 'Header_Color'],[.5, 'Odd_Row_Color'],[1, 'Even_Row_Color']]
colorscale = [[0, '#4a9eec'],[.5, 'd2d2d2'],[1, '#ffffff']]

table = FF.create_table(ranked_companies, colorscale=colorscale)
plot_url = py.plot(table, filename='color_table')
```

```python
df2 = mat_leave [[3,1]]
df2.columns = ['Industry', 'Paid_Weeks']
industry_rank = df2.groupby('Industry').mean().sort_values(by='Paid_Weeks',ascending=False).round(decimals=1).reset_index()
```

```python
industry_rank.head()
```

```python

x = industry_rank['Industry']
y = industry_rank['Paid_Weeks']

data = [
    go.Bar(
        x=x,
        y=y,
        marker=dict(
            color='rgb(158,202,225)',
            line=dict(
                color='rgb(8,48,107)',
                width=1.5
            ),
        ),
        opacity=0.6
    )
]
layout = go.Layout(
      title='Average Paid Maternity Leave by Industry',
    xaxis=dict(
        title='Industry'
    ),
    yaxis=dict(
        title='# of paid weeks off'
    ),
    
    annotations=[
        dict(
            x=xi,
            y=yi,
            text=str(yi),
            xanchor='center',
            yanchor='bottom',
            showarrow=False,
        ) for xi, yi in zip(x, y)]
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='mat_leave_industry_cut')
```

```python
clean_data = PerDiem2015[[0,7]] #had to index using numbers, syntax should also allow calling column names, 'Per Diem' kicked back error, though
clean_data.head(10)
```

```python
country_group = clean_data.groupby('Country')
perDiem_totals = country_group.sum() #necessary to pivot/sum data, could be lumped into above line
perDiem_totals.head()

top10 = clean_data.head(10).groupby('Country').first() #removed the duplicate countries

my_plot = top10.plot(kind='bar',title="Top 8 countries with largest state department per diem",figsize=(9, 7))
my_plot.set_ylabel("Per Diem")
```

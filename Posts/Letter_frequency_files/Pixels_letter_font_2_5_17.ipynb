## Helvetica (single font) pixel usage

```python
from operator import itemgetter
from PIL import Image, ImageDraw, ImageFont
```

## fonts

Roboto

Comic Sans MS  
Consolas

Times New Roman  

```python
# Make a lowercase + uppercase alphabet.
alphabet = 'abcdefghijklmnopqrstuvwxyz'
alphabet += ''.join(map(str.upper, alphabet))


# We'll use Helvetica in big type.
TimesNewRoman = ImageFont.truetype('../English/Times New Roman/Times New Roman.ttf', 100)


def draw_letter(letter, save=True):
    img = Image.new('RGB', (100, 100), 'white')

    draw = ImageDraw.Draw(img)
    draw.text((0,0), letter, font=TimesNewRoman, fill='#000000')

    if save:
        img.save("{}.png".format(letter), 'PNG')

    return img


def count_black_pixels(img):
    pixels = list(img.getdata())
    l = list(filter(lambda rgb: sum(rgb) == 0, pixels))
    return len(l)


if __name__ == '__main__':
    counts = [
        (letter, count_black_pixels(draw_letter(letter)))
        for letter in alphabet
    ]

    TimesNewRoman = (counts)
    #print sorted(counts, key=itemgetter(1), reverse=True)
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(TimesNewRoman)
df['Font'] = 'Times_New_Roman'
df.columns = ['Letter', 'Pixels', 'Font']
df.to_csv('../fonts/Times_New_Roman.csv')
df.head(50)
```
